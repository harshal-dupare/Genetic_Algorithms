```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as img
```

### simulates the flooding of water into a region where only horizontal and vertical flow of water is allowed and flooding starts from the edges

```python

def iflow(x,n,m,k):
    
    for i in range(m):
        if x[k-1,i] < 200 and x[k,i] < 1 and x[k-1,i] > 20:
            x[k,i] = 127
            
    for i in range(m):
        if x[n-k,i] < 200 and x[n-k-1,i] < 1 and x[n-k,i] > 20:
            x[n-k-1,i] = 127
            
    for i in range(n):
        if x[i,k-1] < 200 and x[i,k] < 1 and x[i,k-1] > 20:
            x[i,k] = 127
            
    for i in range(n):
        if x[i,m-k] < 200 and x[i,m-k-1] < 1 and x[i,m-k] > 20:
            x[i,m-k-1] = 127
    
    return


def initflow(x,n,m):
    
    for i in range(m):
        if x[0,i] < 1:
            x[0,i] = 127
            
    for i in range(m):
        if x[n-1,i] < 1:
            x[n-1,i] = 127
            
    for i in range(n):
        if x[i,0] < 1:
            x[i,0] = 127
            
    for i in range(n):
        if x[i,m-1] < 1:
            x[i,m-1] = 127
    
    return
```

```python
n=100
m=100
x = 254*np.uint8((np.random.random((n,m))>0.8))
initflow(x,n,m)
plt.imshow(x,cmap='Blues') 
plt.show()
for k in range((n-2//2)):
    iflow(x,n,m,k+1)
    plt.imshow(x,cmap='Blues') 
    plt.show()
plt.show()
plt.figure(figsize=(2,2))
```
