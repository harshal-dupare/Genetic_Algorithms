```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as img
```

### yellow being the warmest cells and blue the coldest 
* we can observe the heat transfer

```python
def Heat_difu(x,n,m):
    y=np.zeros((n,m))
    for i in range(n):
        for j in range(m):
            y[i,j]=((x[(i+1)%n,(j)%m]+x[(i-1)%n,(j)%m]+x[(i)%n,(j+1)%m]+x[(i)%n,(j-1)%m])//4).astype(np.uint8)
            
    return y



n=100
m=100
x = 255*np.int8((np.random.random((n,m))>0.1))
for i in range(50):
    x=Heat_difu(x,n,m)
    plt.imshow(x,cmap='plasma_r') 
    plt.show()
plt.show()
plt.figure(figsize=(2,2))
```
