```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as img
```

```python
def CGL(x,n,m):
    y=np.zeros((n,m))
    for i in range(n):
        for j in range(m):
            c=((x[(i+1)%n,(j)%m]+x[(i-1)%n,(j)%m]+x[(i)%n,(j+1)%m]+x[(i)%n,(j-1)%m]))
            y[i,j]=x[i,j]
            if bool(x[i,j]):
                if c < 2:
                    y[i,j]=0
                if c is 4:
                    y[i,j]=0
            else :
                if c is 3:
                    y[i,j]=1
                    
    return y

n=100
m=100
x = np.int8((np.random.random((n,m))>0.5))
for i in range(25):
    plt.imshow(x)
    plt.show() 
    x=CGL(x,n,m)
plt.show()
plt.figure(figsize=(2,2))
```
