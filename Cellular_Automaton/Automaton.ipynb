```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as img
```

```python
def up(x,n,m,p=0.3):
    y= np.int8(np.random.random((n,m))>1)
    
    for i in range(n):
        for j in range(m):
            c=0
            c+=x[(i+1)%n,(j)%m]+x[(i-1)%n,(j)%m]+x[(i)%n,(j+1)%m]+x[(i)%n,(j-1)%m]
            y[i,j]=x[i,j]
            if c is 3 and x[i,j] is 0:
                y[i,j]=1
            if c is 2 and x[i,j] is 0:
                y[i,j]=1
            if c < 2:
                y[i,j]=int(np.random.rand()<p)
    return y
    

def upd(x,n,m):
    y= np.int8(np.random.random((n,m))>1)
    for i in range(n):
        for j in range(m):
            c=0
            c+=x[(i+1)%n,(j)%m]+x[(i-1)%n,(j)%m]+x[(i)%n,(j+1)%m]+x[(i)%n,(j-1)%m]
            if c > 2 :
                y[i,j]=1
            if c is 0:
                y[i,j]=int(np.random.rand()>0.5)
    return y

def upda(x,n,m):
    y= np.int8(np.random.random((n,m))>1)
    for i in range(n):
        for j in range(m):
            c=0
            c+=x[(i+1)%n,(j)%m]+x[(i-1)%n,(j)%m]+x[(i)%n,(j+1)%m]+x[(i)%n,(j-1)%m]
            if c is 4 :
                y[i,j]=0
            if c is 1 :
                y[i,j]=1
            else :
                y[i,j]=x[i,j]
    return y
    
n=100
m=100
x = np.int8(np.random.random((n,m))>0.8)
for i in range(50):
    plt.imshow(x, cmap="gray")
    plt.show() 
    x=up(x,n,m,p=float(0.002/((i%5)+1)))
plt.show()
plt.figure(figsize=(2,2))
```
